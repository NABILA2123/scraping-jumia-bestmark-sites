```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd 
import re
```

```python
responses = []
for i in range(21):
    url ='https://www.jumia.ma/smartphones/?page='+str(i)+'#catalog-listing'
    responses.append(requests.get(url))
print(responses)
```

```python
soups=[]
for i in range(len(responses)):
    soups.append(BeautifulSoup(responses[i].text,'html.parser'))
```

```python
allnames=[]
for soup in soups:
    names = soup.find_all('h3' , {'class':'name'})
    allnames.append(names)
print(allnames)
print(len(allnames))
print(len(allnames[0]))
```

```python
Names=[]
for name in allnames:
    for naame in name:
        Names.append(naame.text.strip())
print(Names)
print(len(Names))
```

```python
cleaned_phones=[]
result=[]
for element in Names:
    cleaned_phones = element.split()[:3]
    result.append(' '.join(cleaned_phones))
print(result)
```

```python
prices=[]
for soup in soups:
    price = soup.findAll('div', class_='prc')
    prices.append(price)
print(prices)
```

```python
Prices=[]
for price in prices:
    for priice in price:
         Prices.append(priice.text.strip())
print(Prices)
print(len(Prices))
```

```python
telephones=[]
for i in range(len(Names)):
        dictio={'name':result[i],'price1':Prices[i]}
        telephones.append(dictio)
print(telephones)
print(len(telephones))
```

```python
with open(r"C:\Users\PC\Downloads\scrapingallpages","w",encoding='utf-8') as file:
    colonnes = ['name','price1']
    writer =  csv.DictWriter(file,colonnes) 
    writer.writeheader()
    for tel in telephones :
        writer.writerow(tel)
```
